```python
# Start from here
from keras import models
from keras import layers
#from keras.layers.normalization import BatchNormalization

model = models.Sequential()
model.add(layers.Conv2D(32, (25, 25),activation='relu', strides=(2, 2), input_shape=(100, 100,1)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64, (4, 4),activation='relu', strides=(1, 1)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128, (3, 3),activation='relu', strides=(1, 1)))
model.add(layers.Flatten())
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
#from keras.utils import to_categorical
train_images = train_image.reshape((-1,100,100,1))
test_images = test_image.reshape((-1,100,100,1))
```

```python
model.compile(optimizer='rmsprop',
              loss='mean_squared_error',
              metrics=['accuracy'])
```

```python
#history = model.fit(train_images, train_label, epochs=20, batch_size=160, validation_split=0.25)
history = model.fit(train_images, train_label, epochs=5, batch_size=160, validation_data=(test_images, test_label))
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

print('Red:    accuracy of training data\nOrange: loss of training data\nBlue:   accuracy of testing data\nGreen:  loss of testing data')
plt.figure(figsize=(15,7))
plt.plot(history.history.get('acc'), color = 'r')
plt.plot(history.history.get('loss'), color = 'orange')
plt.plot(history.history.get('val_acc'), color = 'b')
plt.plot(history.history.get('val_loss'), color = 'green')
plt.xlabel('Epochs')
plt.ylabel('The value of accuracy or loss')

plt.show()

```
